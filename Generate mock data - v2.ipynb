```python
import pandas as pd
import numpy as np
import datetime
from random import seed
from random import random
import random as r
import calendar
```

```python
# Next we will design a dictionary 'products' to hold product name and prices. In this case, we only uses generic product names. And the price is generated with specified seeds for repeatability
# weight impact purchase frequency (later will be used to generate data frame)

category_list=['Home & Garden','Health & Personal Care', 'Major Appliances', 'Grocery & Gourmet Food', 'Musical Instrument',
               'Books & eBooks', 'Office Supplies', 'Toys & Games', 'Electronics']
def generate_category(n):
    if n<=10: c=0; m=0.38
    elif n>10 and n<=25 : c=1; m=0.25
    elif n>25 and n<=32: c=2; m=0.35
    elif n>32 and n<=50: c=3; m=0.28
    elif n>50 and n<=56: c=4; m=0.4
    elif n>56 and n<=70: c=5; m=0.7
    elif n>70 and n<=80: c=6; m=0.33
    elif n>80 and n<=95: c=7; m=0.37
    else: c=8 ; m=0.42
    result=[category_list[c],m]
    return result

def generate_price(n):
    seed(n)
    if n<=10: c=200
    elif n>10 and n<=25 : c=100
    elif n>25 and n<=32: c=1000
    elif n>32 and n<=50: c=30
    elif n>50 and n<=56: c=600
    elif n>56 and n<=70: c=100
    elif n>70 and n<=80: c=50
    elif n>80 and n<=95: c=100
    else: c=1000
    return round(random()*c,2)
    
product_names=[]
product_prices=[]
weights=[]
categories=[]
margins=[]
i=1
while i<101:
        product='Product '+str(i)
        product_names.append(product)
        price=generate_price(i)
        product_prices.append(price)
        weight=r.randint(1,100)
        weights.append(weight)
        category=generate_category(i)[0]
        categories.append(category)
        margin=generate_category(i)[1]
        margins.append(margin)
        i+=1
print(product_names) 
print(product_prices)
print(weights)
print(categories)
print(margins)
```

```python
customer_list=[]
customer_zip_code=[]
customer_ID=100102
uszips=pd.read_csv('uszips.csv')
zip_list=uszips['zip']
j=1
while j < 100000:
    customer_list.append(customer_ID)
    customer_ID+=1
    zipcode=r.choices(zip_list)[0]
    customer_zip_code.append(zipcode)
    j+=1
print(customer_list)
print(customer_zip_code)

customers=dict(zip(customer_list,customer_zip_code))
customers
    
```

```python
# Now we have our dictionary
# this a dictionary with product_names as key, and each value is a short list that holds price and weight(impact purchasing frequency)
pw= list(map(list,zip(product_prices,margins,weights,categories)))
products=dict(zip(product_names,pw))

print(pw)
print(products)
```

```python
# Here is how we generate specific number of rows in each month
# This dataset has built in seasonality in purchase numbers (refer to order_amount)
# This dataset has built in the variance of purchasing frequency based on product (refer to weights)
# generated purchase zip code using a random.choices. List of zip codes in the US is downloaded from: https://simplemaps.com/data/us-zips
# Order date has date and specific time. 
# this is also take into the cosnideration of frequency of order time (8 pm vs. 8 am for instance)
# quantity in each order is generated based on geometric distribution. 

# these are the column names for our dataframe
columns=['Order ID', 'Consumer ID','Product Name','Category','Quantity','Price Each','Profit','Order Date','Purchase Zip Code', 
         'Delayed Shipping','Consumer Rating','Return Status']
# define the initial order_id.
order_id =143679

def generate_random_time(month):
    day_range=calendar.monthrange(2020,month)[1] # day range within each month (ie., 28 days in Feb, 31 days in Jan, etc)
    random_day=r.randint(1,day_range)
    # datetime.datetime() takes in year,month, day, hour, min, sec
    if r.random() <0.5:
        date=datetime.datetime(2020,month,random_day, 12, 0)
    else:
        date=datetime.datetime(2020,month,random_day, 12, 0)
    time_offset = np.random.normal(loc=0, scale=180)
    
    final_date=date+datetime.timedelta(minutes=time_offset)
    return final_date.strftime("%m/%d/%Y %H:%M")

def generate_delayed_arrival():
    if r.random()<0.1: delayed=1
    else:delayed=0
    return delayed

def generate_rating(delayed):
    if delayed==1: rating=r.choices([1,2,3,4,5],weights=[10,20,50,20,10])[0]
    else: rating=r.choices([1,2,3,4,5],weights=[1,20,50,40,30])[0]
    return rating

def generate_returned(rating):
    if rating>=4:
        if r.random()<0.05: returned=1
        else: returned=0
    elif rating==3:
        if r.random()<0.2: returned=1
        else:returned=0
    else:
        if r.random()<0.5: returned=1
        else:returned=0
    return returned

for month_value in range(1,13):
    if month_value<=10:
        order_amount=int(np.random.normal(loc=12000, scale=4000)) # loc is the mean, scale is the std/spread
        discount=0.05
    if month_value==11:
        order_amount=int(np.random.normal(loc=20000, scale=3000))
        discount=0.1
        # slightly higher records (higher mean than other months)
    if month_value ==12:
        order_amount=int(np.random.normal(loc=26000, scale=3000))
        discount=0.2
        # highest number of purchase records (highest mean)
    
    df=pd.DataFrame(columns=columns)
    for i in range(order_amount):
        product=r.choices(product_names,weights=weights)[0] # random.choices() takes in a list and the correponding frequency weight. You can define how many it would choose by specifying k=n. k is 1 by default. The output is a list, so we will need to use index 0 to extract the content at index 0.
        category=products[product][3]
        price=products[product][0]
        mar=products[product][1] # margin 
        profit=price*mar*(1-discount)
        quantity_ordered=np.random.geometric(p=1.0-(1.0/price),size=1)
        customer=r.choices(customer_list)[0]
        zip_code=customers[customer]
        date=generate_random_time(month_value)
        delayed=generate_delayed_arrival()
        rating=generate_rating(delayed)
        returned=generate_returned(delayed)
        df.loc[i]=[order_id,customer,product,category,quantity_ordered[0],price,profit,date,zip_code, delayed, rating, returned]
        order_id+=1
    month_name=calendar.month_name[month_value]
    print(month_name + " finished!")
    df.to_csv(str(month_value)+"."+f"{month_name}_data.csv")
```
